```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
voc_size=10000
```

```python
one_hot_rep=[one_hot(words,voc_size) for words in sent]
```

```python
one_hot_rep
```

```python
# word embedding rep

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences 
from tensorflow.keras.models import Sequential


```

```python
import numpy as np 

```

```python
sent_len=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len)
```

```python
print(embedded_docs)
```

```python
#feature representation 
dim=10
```

```python
model = Sequential()

# Add the Embedding layer
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=sent_len))

# Manually build the model
model.build(input_shape=(None, sent_len))


# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
